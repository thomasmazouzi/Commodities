### Question 2

Assuming that the US 10y yield represents the risk free rate, compute for gold and silver the sharpe ratio of a passive (long-only) investment for each of the available years and over all of the available history.

```python
import numpy as np

def sharpe_ratio(return_series, rf_series, N):
    excess_returns = return_series - rf_series
    mean_excess_return = excess_returns.mean() * N
    volatility = excess_returns.std() * np.sqrt(N)
    return mean_excess_return / volatility


# Sharpe ratio by year
sharpes_by_year = returns.groupby(returns.index.year).apply(
    lambda x: pd.Series({
        'Gold': sharpe_ratio(x['Gold'], x['USD 10y yield'], len(x)),
        'Silver': sharpe_ratio(x['Silver'], x['USD 10y yield'], len(x)) #dynamic N to avoid problem with the last year
    })
)

# Total Sharpe ratio
sharpes_global = pd.Series({
    'Gold': sharpe_ratio(returns['Gold'], returns['USD 10y yield'], len(returns)),
    'Silver': sharpe_ratio(returns['Silver'], returns['USD 10y yield'], len(returns))
})
```

```python
# Plot Sharpe Ratios by year
sharpes_by_year.plot.bar(figsize=(10, 6))
plt.title("Sharpe Ratios by Year for Gold and Silver")
plt.ylabel("Sharpe Ratio")
plt.xlabel("Year")
plt.show()
```

```python
import seaborn as sns
from statsmodels.tsa.stattools import adfuller

def adf_test(series):
    series = series.dropna()  # Without NaN value
    if len(series) < 10: 
        return {"Test Statistic": np.nan, "p-value": np.nan, "Stationary": np.nan}
    result = adfuller(series, autolag="AIC")
    return {
        "Test Statistic": result[0],
        "p-value": result[1],
        "Critical Value (1%)": result[4]["1%"],
        "Critical Value (5%)": result[4]["5%"],
        "Critical Value (10%)": result[4]["10%"],
        "Stationary": result[1] < 0.05,
    }

# Apply the test for each commodity
adf_results_returns = {col: adf_test(commodity[col]) for col in commodity.columns}

adf_df_returns = pd.DataFrame(adf_results_returns).T

plt.figure(figsize=(10, 6))
sns.barplot(x=adf_df_returns.index, y=adf_df_returns["Test Statistic"], palette="coolwarm")
plt.axhline(y=adf_df_returns["Critical Value (5%)"].mean(), color="red", linestyle="dashed", label="Critical Value (5%)")
plt.ylabel("ADF Test Statistic")
plt.title("ADF Test Results for Price")
plt.xticks(rotation=90)
plt.legend()
plt.show()

adf_df_returns
```

### Question 4

For each commodity, compute its 21 day - rolling realized volatility. Display in scatterplots realized vols vs prices. Comment.  



```python
# Calculate 21-day rolling realized volatility for each commodity
rolling_vol = returns_commodity.rolling(window=21).std() * np.sqrt(252)  # Annualized

# Create a function to plot volatility vs price for each commodity
def plot_vol_vs_price(commodity_name):
    plt.scatter(commodity[commodity_name], rolling_vol[commodity_name], alpha=0.5)
    plt.title(f'Price vs Realized Volatility: {commodity_name}')
    plt.xlabel('Price')
    plt.ylabel('21-day Rolling Realized Volatility (Annualized)')
    plt.grid(True, alpha=0.3)
    # Add regression line
    valid_data = ~(np.isnan(commodity[commodity_name]) | np.isnan(rolling_vol[commodity_name]))
    if sum(valid_data) > 2:  # Need at least two points for regression
        x = commodity[commodity_name][valid_data]
        y = rolling_vol[commodity_name][valid_data]
        z = np.polyfit(x, y, 1)
        p = np.poly1d(z)
        plt.plot(x, p(x), 'r--', alpha=0.8)
        # Calculate correlation
        corr = np.corrcoef(x, y)[0, 1]
        plt.annotate(f'Correlation: {corr:.2f}', xy=(0.05, 0.95), xycoords='axes fraction', fontsize=10, color='blue')
    plt.tight_layout()
    plt.show()

# Plot for each commodity (first 4 for brevity)
for commodity_name in list(commodity.columns[:4]):
    plot_vol_vs_price(commodity_name)
```

```python
# Create a heatmap to visualize the relationship between volatility and prices for all commodities
correlations = []
commodities = []

for commodity_name in commodity.columns:
    valid_data = ~(np.isnan(commodity[commodity_name]) | np.isnan(rolling_vol[commodity_name]))
    if sum(valid_data) > 2:  # Need at least two points for correlation
        x = commodity[commodity_name][valid_data]
        y = rolling_vol[commodity_name][valid_data]
        corr = np.corrcoef(x, y)[0, 1]
        correlations.append(corr)
        commodities.append(commodity_name)

# Create a DataFrame for the correlations
corr_df = pd.DataFrame({'Commodity': commodities, 'Correlation': correlations})
corr_df = corr_df.sort_values('Correlation')

# Plot the correlations
plt.figure(figsize=(10, 6))
sns.barplot(x='Correlation', y='Commodity', data=corr_df, palette='coolwarm')
plt.title('Correlation between Price and Volatility for Each Commodity')
plt.axvline(x=0, color='black', linestyle='--')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

# Comment on the findings
print("Analysis of Price vs. Volatility Relationships:")
print("1. Most commodities show some level of correlation between price and volatility")
print("2. Positive correlation suggests that higher prices are associated with higher volatility")
print("3. Negative correlation suggests that lower prices are associated with higher volatility (volatility clustering during bearish periods)")
print("4. This relationship is commonly observed in commodities and is often referred to as the 'leverage effect'")
```
