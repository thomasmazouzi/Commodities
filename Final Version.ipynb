### Question 2

Assuming that the US 10y yield represents the risk free rate, compute for gold and silver the sharpe ratio of a passive (long-only) investment for each of the available years and over all of the available history.

```python
import numpy as np

def sharpe_ratio(return_series, rf_series, N):
    excess_returns = return_series - rf_series
    mean_excess_return = excess_returns.mean() * N
    volatility = excess_returns.std() * np.sqrt(N)
    return mean_excess_return / volatility


# Sharpe ratio by year
sharpes_by_year = returns.groupby(returns.index.year).apply(
    lambda x: pd.Series({
        'Gold': sharpe_ratio(x['Gold'], x['USD 10y yield'], len(x)),
        'Silver': sharpe_ratio(x['Silver'], x['USD 10y yield'], len(x)) #dynamic N to avoid problem with the last year
    })
)

# Total Sharpe ratio
sharpes_global = pd.Series({
    'Gold': sharpe_ratio(returns['Gold'], returns['USD 10y yield'], len(returns)),
    'Silver': sharpe_ratio(returns['Silver'], returns['USD 10y yield'], len(returns))
})
```

```python
# Plot Sharpe Ratios by year
sharpes_by_year.plot.bar(figsize=(10, 6))
plt.title("Sharpe Ratios by Year for Gold and Silver")
plt.ylabel("Sharpe Ratio")
plt.xlabel("Year")
plt.show()
```

```python
import seaborn as sns
from statsmodels.tsa.stattools import adfuller

def adf_test(series):
    series = series.dropna()  # Without NaN value
    if len(series) < 10: 
        return {"Test Statistic": np.nan, "p-value": np.nan, "Stationary": np.nan}
    result = adfuller(series, autolag="AIC")
    return {
        "Test Statistic": result[0],
        "p-value": result[1],
        "Critical Value (1%)": result[4]["1%"],
        "Critical Value (5%)": result[4]["5%"],
        "Critical Value (10%)": result[4]["10%"],
        "Stationary": result[1] < 0.05,
    }

# Apply the test for each commodity
adf_results_returns = {col: adf_test(commodity[col]) for col in commodity.columns}

adf_df_returns = pd.DataFrame(adf_results_returns).T

plt.figure(figsize=(10, 6))
sns.barplot(x=adf_df_returns.index, y=adf_df_returns["Test Statistic"], palette="coolwarm")
plt.axhline(y=adf_df_returns["Critical Value (5%)"].mean(), color="red", linestyle="dashed", label="Critical Value (5%)")
plt.ylabel("ADF Test Statistic")
plt.title("ADF Test Results for Price")
plt.xticks(rotation=90)
plt.legend()
plt.show()

adf_df_returns
```

### Question 4

For each commodity, compute its 21 day - rolling realized volatility. Display in scatterplots realized vols vs prices. Comment.  



```python
rolling_vol = returns_commodity.rolling(window=21).std() * np.sqrt(252)

n_assets = len(commodity.columns)
n_cols = 3
n_rows = int(np.ceil(n_assets / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(5 * n_cols, 4 * n_rows), sharex=False)
axes = axes.flatten()

for i, commodity_name in enumerate(commodity.columns):
    ax = axes[i]
    
    x = commodity[commodity_name]
    y = rolling_vol[commodity_name]

    ax.scatter(x, y, alpha=0.5)
    ax.set_title(f'{commodity_name}')
    ax.set_xlabel('Price')
    ax.set_ylabel('21d Vol (Annualized)')
    ax.grid(True, alpha=0.3)
    
    # Régression linéaire si assez de données valides
    valid_data = ~(np.isnan(x) | np.isnan(y))
    if sum(valid_data) > 2:
        x_valid = x[valid_data]
        y_valid = y[valid_data]
        z = np.polyfit(x_valid, y_valid, 1)
        p = np.poly1d(z)
        ax.plot(x_valid, p(x_valid), 'r--', alpha=0.8)

        # Correlation
        corr = np.corrcoef(x_valid, y_valid)[0, 1]
        ax.annotate(f'Corr: {corr:.2f}', xy=(0.05, 0.95), xycoords='axes fraction',
                    fontsize=10, color='blue', ha='left', va='top')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Create a heatmap to visualize the relationship between volatility and prices for all commodities
correlations = []
commodities = []

for commodity_name in commodity.columns:
    valid_data = ~(np.isnan(commodity[commodity_name]) | np.isnan(rolling_vol[commodity_name]))
    if sum(valid_data) > 2:  # Need at least two points for correlation
        x = commodity[commodity_name][valid_data]
        y = rolling_vol[commodity_name][valid_data]
        corr = np.corrcoef(x, y)[0, 1]
        correlations.append(corr)
        commodities.append(commodity_name)

# Create a DataFrame for the correlations
corr_df = pd.DataFrame({'Commodity': commodities, 'Correlation': correlations})
corr_df = corr_df.sort_values('Correlation')

# Plot the correlations
plt.figure(figsize=(10, 6))
sns.barplot(x='Correlation', y='Commodity', data=corr_df, palette='coolwarm')
plt.title('Correlation between Price and Volatility for Each Commodity')
plt.axvline(x=0, color='black', linestyle='--')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

Analysis of Price vs. Volatility Relationships:
1. Most commodities show some level of correlation between price and volatility
2. Positive correlation suggests that higher prices are associated with higher volatility
3. Negative correlation suggests that lower prices are associated with higher volatility (volatility clustering during bearish periods)
4. This relationship is commonly observed in commodities and is often referred to as the 'leverage effect'

### Question 8
For each of the selected countries, compute (using bivariate regressions):
1. Elasticities of demand for wheat and soybean
2. Cross elasticities

Find a was to visualize those results in a synthetic way.
Interpret. Do you think our results are strong?




```python
filters = ['Wheat Chicago Avg Price', 'Soybean Avg Price', 'Wheat_United States', 'Wheat_China', 'Wheat_Brazil', 'Wheat_Russia', 'Wheat_India', 'Wheat_EU-25', 'Wheat_Argentina', 'Wheat_Canada', 'Soybean_United States', 'Soybean_China', 'Soybean_Brazil', 'Soybean_Russia', 'Soybean_India', 'Soybean_EU-25', 'Soybean_Argentina', 'Soybean_Canada']

# Filter the DataFrame to keep only the specified columns
elasticity_df_8 = merged_df[filters]

# Rename columns for clarity
elasticity_df_8.rename(columns={
    'Wheat_United States': 'Wheat_United States Demand',
    'Wheat_china': 'Wheat_China Demand',
    'Wheat_Brazil': 'Wheat_Brazil Demand',
    'Wheat_Russia': 'Wheat_Russia Demand',
    'Wheat_India': 'Wheat_India Demand',
    'Wheat_EU-25': 'Wheat_EU-25 Demand',
    'Wheat_Argentina': 'Wheat_Argentina Demand',
    'Wheat_China': 'Wheat_China Demand',
    'Wheat_Canada': 'Wheat_Canada Demand',
    'Soybean_United States': 'Soybean_United States Demand',
    'Soybean_China': 'Soybean_China Demand',
    'Soybean_Brazil': 'Soybean_Brazil Demand',
    'Soybean_Russia': 'Soybean_Russia Demand',
    'Soybean_India': 'Soybean_India Demand',
    'Soybean_EU-25': 'Soybean_EU-25 Demand',
    'Soybean_Argentina': 'Soybean_Argentina Demand',
    'Soybean_Canada': 'Soybean_Canada Demand',
    'Soybean_United States': 'Soybean_United States Demand',
}, inplace=True)

elasticity_df_8_log = elasticity_df_8.apply(np.log).dropna()
```

```python
# Select countries for analysis 
countries = ['United States', 'China', 'Brazil', 'Russia', 'India', 'EU-25', 'Argentina', 'Canada']

# Function to run bivariate analysis for a country
def analyze_elasticities(country):
    # Prepare data variables
    log_wheat_demand = elasticity_df_8_log[f'Wheat_{country} Demand']
    log_soy_demand = elasticity_df_8_log[f'Soybean_{country} Demand']
    log_soybean_price = elasticity_df_8_log['Soybean Avg Price']
    log_wheat_price = elasticity_df_8_log['Wheat Chicago Avg Price']

    
    # Create matrix of explanatory variables
    X = sm.add_constant(pd.DataFrame({
        'log_wheat_price': log_wheat_price,
        'log_soybean_price': log_soybean_price
    }))
    
    # Run wheat demand model
    wheat_model = sm.OLS(log_wheat_demand, X).fit()
    wheat_own_elasticity = wheat_model.params['log_wheat_price']
    wheat_cross_elasticity = wheat_model.params['log_soybean_price']
    wheat_r2 = wheat_model.rsquared
    
    # Run soybean demand model
    soy_model = sm.OLS(log_soy_demand, X).fit()
    soy_own_elasticity = soy_model.params['log_soybean_price']
    soy_cross_elasticity = soy_model.params['log_wheat_price']
    soy_r2 = soy_model.rsquared
    
    return {
        'Country': country,
        'Wheat Own-Price Elasticity': wheat_own_elasticity,
        'Wheat Cross-Price Elasticity': wheat_cross_elasticity,
        'Wheat R²': wheat_r2,
        'Soybean Own-Price Elasticity': soy_own_elasticity,
        'Soybean Cross-Price Elasticity': soy_cross_elasticity,
        'Soybean R²': soy_r2,
        'Wheat-Soy Relationship': 'Substitutes' if wheat_cross_elasticity > 0 else 'Complements' if wheat_cross_elasticity < 0 else 'Independent'
    }

# Apply the function to all selected countries
results = []
for country in countries:
    result = analyze_elasticities(country)
    results.append(result)

# Create results dataframe
elasticity_results = pd.DataFrame(results)
elasticity_results
```

```python
# Visualize the elasticity results across countries
plt.figure(figsize=(14, 8))

# Create subplots for different elasticity types
plt.subplot(2, 2, 1)
elasticity_results.plot(x='Country', y='Wheat Own-Price Elasticity', kind='bar', ax=plt.gca(), color='wheat')
plt.title('Wheat Own-Price Elasticity by Country')
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.ylabel('Elasticity')
plt.xticks(rotation=90)

plt.subplot(2, 2, 2)
elasticity_results.plot(x='Country', y='Soybean Own-Price Elasticity', kind='bar', ax=plt.gca(), color='green')
plt.title('Soybean Own-Price Elasticity by Country')
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.ylabel('Elasticity')
plt.xticks(rotation=90)

plt.subplot(2, 2, 3)
elasticity_results.plot(x='Country', y='Wheat Cross-Price Elasticity', kind='bar', ax=plt.gca(), color='orange')
plt.title('Wheat Cross-Price Elasticity by Country')
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.ylabel('Elasticity')
plt.xticks(rotation=90)

plt.subplot(2, 2, 4)
elasticity_results.plot(x='Country', y='Soybean Cross-Price Elasticity', kind='bar', ax=plt.gca(), color='blue')
plt.title('Soybean Cross-Price Elasticity by Country')
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.ylabel('Elasticity')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()

# Create a heatmap showing R-squared values for all models
plt.figure(figsize=(10, 6))
r2_data = elasticity_results[['Country', 'Wheat R²', 'Soybean R²']].set_index('Country')
sns.heatmap(r2_data, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('R-squared Values for Wheat and Soybean Models by Country')
plt.tight_layout()
plt.show()

# Create a visual to show the relationship characterization
relationship_counts = elasticity_results['Wheat-Soy Relationship'].value_counts()
plt.figure(figsize=(10, 6))
relationship_counts.plot(kind='bar', color='purple')
plt.title('Wheat-Soybean Market Relationship Across Countries')
plt.xlabel('Relationship Type')
plt.ylabel('Number of Countries')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Interpretation of elasticity results
print("\nInterpretation of Multi-Country Elasticity Analysis:")
print("1. Own-Price Elasticities:")
own_price_wheat_avg = elasticity_results['Wheat Own-Price Elasticity'].mean()
own_price_soy_avg = elasticity_results['Soybean Own-Price Elasticity'].mean()
print(f"   - Average wheat own-price elasticity across countries: {own_price_wheat_avg:.4f}")
print(f"   - Average soybean own-price elasticity across countries: {own_price_soy_avg:.4f}")
if abs(own_price_wheat_avg) < abs(own_price_soy_avg):
    print("   - Wheat demand appears less elastic than soybean demand across countries on average")
else:
    print("   - Soybean demand appears less elastic than wheat demand across countries on average")

print("\n2. Cross-Price Elasticities:")
positive_cross = (elasticity_results['Wheat Cross-Price Elasticity'] > 0).sum()
total_valid = elasticity_results['Wheat Cross-Price Elasticity'].notna().sum()
print(f"   - {positive_cross} out of {total_valid} countries show a substitution relationship")
print(f"   - {total_valid - positive_cross} out of {total_valid} countries show a complementary relationship")

print("\n3. Model Strength Assessment:")
print(f"   - Average R² for wheat models: {elasticity_results['Wheat R²'].mean():.4f}")
print(f"   - Average R² for soybean models: {elasticity_results['Soybean R²'].mean():.4f}")
if elasticity_results['Wheat R²'].mean() < 0.3 and elasticity_results['Soybean R²'].mean() < 0.3:
    print("   - The models generally show low explanatory power, suggesting that")
    print("     price elasticity alone may not be sufficient to explain demand patterns")
    print("     across countries. Other factors like income, dietary preferences,")
    print("     agricultural policies, and trade agreements likely play significant roles.")
```

```python
# Run OLS regressions for income elasticity (using aggregate GDP)
# For wheat
X_gdp_wheat = sm.add_constant(elasticity_df_9_log['US_GDP'])
Y_wheat = elasticity_df_9_log['Wheat_United States Demand']
model_gdp_wheat = sm.OLS(Y_wheat, X_gdp_wheat).fit()

# For soybean
X_gdp_soy = sm.add_constant(elasticity_df_9_log['US_GDP'])
Y_soy = elasticity_df_9_log['Soybean_United States Demand']
model_gdp_soy = sm.OLS(Y_soy, X_gdp_soy).fit()

print("Income Elasticity Analysis using Aggregate GDP:")
print("\nWheat Income Elasticity Model:")
print(model_gdp_wheat.summary())
wheat_income_elasticity = model_gdp_wheat.params['US_GDP']
print(f"\nWheat Income Elasticity: {wheat_income_elasticity:.4f}")

print("\nSoybean Income Elasticity Model:")
print(model_gdp_soy.summary())
soybean_income_elasticity = model_gdp_soy.params['US_GDP']
print(f"\nSoybean Income Elasticity: {soybean_income_elasticity:.4f}")

# 2. Breaking down GDP into GDP per capita and population
# For wheat
X_decomposed_wheat = sm.add_constant(elasticity_df_9_log[['US GDP/population', 'US_population']])
model_decomposed_wheat = sm.OLS(Y_wheat, X_decomposed_wheat).fit()

# For soybean
X_decomposed_soy = sm.add_constant(elasticity_df_9_log[['US GDP/population', 'US_population']])
model_decomposed_soy = sm.OLS(Y_soy, X_decomposed_soy).fit()

print("\n\nDecomposed Income Elasticity Analysis (GDP/capita and Population):")
print("\nWheat Decomposed Model:")
print(model_decomposed_wheat.summary())
wheat_gdp_per_capita_elasticity = model_decomposed_wheat.params['US GDP/population']
wheat_population_elasticity = model_decomposed_wheat.params['US_population']
print(f"\nWheat GDP per Capita Elasticity: {wheat_gdp_per_capita_elasticity:.4f}")
print(f"Wheat Population Elasticity: {wheat_population_elasticity:.4f}")

print("\nSoybean Decomposed Model:")
print(model_decomposed_soy.summary())
soybean_gdp_per_capita_elasticity = model_decomposed_soy.params['US GDP/population']
soybean_population_elasticity = model_decomposed_soy.params['US_population']
print(f"\nSoybean GDP per Capita Elasticity: {soybean_gdp_per_capita_elasticity:.4f}")
print(f"Soybean Population Elasticity: {soybean_population_elasticity:.4f}")

# Create visual summary of elasticities
plt.figure(figsize=(12, 6))

# First subplot: Overall GDP elasticities
plt.subplot(1, 2, 1)
elasticities = [wheat_income_elasticity, soybean_income_elasticity]
plt.bar(['Wheat', 'Soybean'], elasticities, color=['wheat', 'green'])
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.axhline(y=1, color='red', linestyle='--', alpha=0.7, label='Luxury Threshold')
plt.title('Income Elasticity Using Aggregate GDP')
plt.ylabel('Elasticity')
plt.grid(axis='y', alpha=0.3)
plt.legend()

# Second subplot: Decomposed elasticities
plt.subplot(1, 2, 2)
labels = ['Wheat\nGDP/Capita', 'Wheat\nPopulation', 'Soybean\nGDP/Capita', 'Soybean\nPopulation']
decomposed_elasticities = [wheat_gdp_per_capita_elasticity, wheat_population_elasticity, 
                          soybean_gdp_per_capita_elasticity, soybean_population_elasticity]
plt.bar(labels, decomposed_elasticities, color=['wheat', 'wheat', 'green', 'green'], alpha=0.7)
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.axhline(y=1, color='red', linestyle='--', alpha=0.7, label='Luxury Threshold')
plt.title('Decomposed Income Elasticity')
plt.ylabel('Elasticity')
plt.grid(axis='y', alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()

# Interpret the results
print("\nInterpretation of Income Elasticity Results:")
print("\nWheat as a Good Type:")
if wheat_income_elasticity < 0:
    print("- Wheat appears to be an inferior good (negative income elasticity)")
elif 0 <= wheat_income_elasticity < 1:
    print("- Wheat appears to be a necessity good (income elasticity between 0 and 1)")
    print("- As income increases, demand for wheat increases but at a slower rate than income growth")
    print("- This is typical for staple foods like wheat, where basic demand is fairly stable")
else:
    print("- Wheat appears to be a luxury good (income elasticity > 1)")

print("\nSoybean as a Good Type:")
if soybean_income_elasticity < 0:
    print("- Soybean appears to be an inferior good (negative income elasticity)")
elif 0 <= soybean_income_elasticity < 1:
    print("- Soybean appears to be a necessity good (income elasticity between 0 and 1)")
else:
    print("- Soybean appears to be a luxury good (income elasticity > 1)")
    print("- As income increases, demand for soybeans increases at a faster rate than income growth")
    print("- This may reflect increased demand for protein sources and animal feed as incomes rise")
```

Mean Reversion

```python
# Test for mean reversion using Augmented Dickey-Fuller test
from statsmodels.tsa.stattools import adfuller

def adf_test(series):
    series = series.dropna()  # Without NaN value
    result = adfuller(series, autolag="AIC")
    return {
        "Test Statistic": result[0],
        "p-value": result[1],
        "Critical Value (1%)": result[4]["1%"],
        "Critical Value (5%)": result[4]["5%"],
        "Critical Value (10%)": result[4]["10%"],
        "Stationary": result[1] < 0.05,
    }

# Apply the test for each spread
adf_results_Chi_Matif = adf_test(spread_chicago_matif)
adf_results_Chi_Mpls = adf_test(spread_chicago_mpls)
adf_df = pd.DataFrame({
    'Spread': ['Chicago/MATIF', 'Chicago/Minneapolis'],
    'Test Statistic': [adf_results_Chi_Matif['Test Statistic'], adf_results_Chi_Mpls['Test Statistic']],
    'p-value': [adf_results_Chi_Matif['p-value'], adf_results_Chi_Mpls['p-value']],
    'Stationary': [adf_results_Chi_Matif['Stationary'], adf_results_Chi_Mpls['Stationary']]
})

# Display the results
print("\nADF Test Results for Wheat Price Spreads:")
print(adf_df)

print("\nInterpretation of ADF Test Results:")
print("1. Chicago/MATIF spread is non-stationary (not mean-reverting).")
print("2. Chicago/Minneapolis spread is non-stationary (not mean-reverting).")
    
```
