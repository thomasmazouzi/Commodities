### Question 2

Assuming that the US 10y yield represents the risk free rate, compute for gold and silver the sharpe ratio of a passive (long-only) investment for each of the available years and over all of the available history.

```python
import numpy as np

def sharpe_ratio(return_series, rf_series, N):
    excess_returns = return_series - rf_series
    mean_excess_return = excess_returns.mean() * N
    volatility = excess_returns.std() * np.sqrt(N)
    return mean_excess_return / volatility


# Sharpe ratio by year
sharpes_by_year = returns.groupby(returns.index.year).apply(
    lambda x: pd.Series({
        'Gold': sharpe_ratio(x['Gold'], x['USD 10y yield'], len(x)),
        'Silver': sharpe_ratio(x['Silver'], x['USD 10y yield'], len(x)) #dynamic N to avoid problem with the last year
    })
)

# Total Sharpe ratio
sharpes_global = pd.Series({
    'Gold': sharpe_ratio(returns['Gold'], returns['USD 10y yield'], len(returns)),
    'Silver': sharpe_ratio(returns['Silver'], returns['USD 10y yield'], len(returns))
})
```

```python
# Plot Sharpe Ratios by year
sharpes_by_year.plot.bar(figsize=(10, 6))
plt.title("Sharpe Ratios by Year for Gold and Silver")
plt.ylabel("Sharpe Ratio")
plt.xlabel("Year")
plt.show()
```

```python
import seaborn as sns
from statsmodels.tsa.stattools import adfuller

def adf_test(series):
    series = series.dropna()  # Without NaN value
    if len(series) < 10: 
        return {"Test Statistic": np.nan, "p-value": np.nan, "Stationary": np.nan}
    result = adfuller(series, autolag="AIC")
    return {
        "Test Statistic": result[0],
        "p-value": result[1],
        "Critical Value (1%)": result[4]["1%"],
        "Critical Value (5%)": result[4]["5%"],
        "Critical Value (10%)": result[4]["10%"],
        "Stationary": result[1] < 0.05,
    }

# Apply the test for each commodity
adf_results_returns = {col: adf_test(commodity[col]) for col in commodity.columns}

adf_df_returns = pd.DataFrame(adf_results_returns).T

plt.figure(figsize=(10, 6))
sns.barplot(x=adf_df_returns.index, y=adf_df_returns["Test Statistic"], palette="coolwarm")
plt.axhline(y=adf_df_returns["Critical Value (5%)"].mean(), color="red", linestyle="dashed", label="Critical Value (5%)")
plt.ylabel("ADF Test Statistic")
plt.title("ADF Test Results for Price")
plt.xticks(rotation=90)
plt.legend()
plt.show()

adf_df_returns
```

```python
demand_data = pd.read_csv("demand.csv")
demand_data['Dates'] = pd.to_datetime(demand_data['Dates'], format="%d/%m/%Y")
demand_data['Year'] = demand_data['Dates'].dt.year + 1  # Décalage si tu veux l’année suivante
demand_data.set_index('Dates', inplace=True)
demand_data['Dates'] = demand_data.index
demand_data
```

```python
annual_prices = commodity[['Wheat Chicago', 'Soybean']].resample('Y').mean()
annual_prices['Year'] = annual_prices.index.year
annual_prices
```

```python
merged_df = pd.merge(demand_data.reset_index(drop=True), annual_prices, on='Year', how='outer')
merged_df.set_index('Dates', inplace=True)
merged_df.rename(columns={'Wheat Chicago': 'Wheat Chicago Avg Price', 'Soybean': 'Soybean Avg Price'}, inplace=True)
elasticity_df = merged_df[[
    'Wheat_United States', 'Wheat Chicago Avg Price',
    'Soybean_United States', 'Soybean Avg Price'
]].copy()

# Suppression des lignes avec des NaN
elasticity_df.dropna(inplace=True)

# Création des colonnes logarithmiques
elasticity_df['Log Wheat Demand US'] =np.log(merged_df['Wheat_United States'])
elasticity_df['Log Wheat Price US'] = np.log(merged_df['Wheat Chicago Avg Price'])
elasticity_df['Log Soybean Demand US'] =np.log(merged_df['Soybean_United States'])
elasticity_df['Log Soybean Price US'] = np.log(merged_df['Soybean Avg Price'])
elasticity_df
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(elasticity_df['Log Wheat Price US'], elasticity_df['Log Wheat Demand US'], alpha=0.7)
plt.title('Log Wheat Price vs Log Wheat Demand (US)')
plt.xlabel('Log Price of Chicago Wheat')
plt.ylabel('Log Demand of Wheat (US)')
plt.grid(True, alpha=0.3)

z = np.polyfit(elasticity_df['Log Wheat Price US'], elasticity_df['Log Wheat Demand US'], 1)
p = np.poly1d(z)
plt.plot(elasticity_df['Log Wheat Price US'], p(elasticity_df['Log Wheat Price US']), "r--", alpha=0.8)
plt.annotate(f'Slope: {z[0]:.3f}', xy=(0.05, 0.95), xycoords='axes fraction')
plt.show()
```

```python
# OLS to estimatice elasticity of US wheat
X = elasticity_df['Log Wheat Price US']
X = sm.add_constant(X)  
Y = elasticity_df['Log Wheat Demand US']

# Fit the model
model = sm.OLS(Y, X).fit()

print(model.summary())
```

Interpretation of Wheat Price Elasticity:

The estimated price elasticity of wheat demand is -0.0580

This indicates inelastic demand: changes in price have a relatively small effect on quantity demanded.
This is typical for staple food products like wheat, which are considered necessities.
R-squared: 0.0260
The model explains 2.60% of the variation in wheat demand.
