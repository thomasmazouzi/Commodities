### Question 2

Assuming that the US 10y yield represents the risk free rate, compute for gold and silver the sharpe ratio of a passive (long-only) investment for each of the available years and over all of the available history.

```python
import numpy as np

def sharpe_ratio(return_series, rf_series, N):
    excess_returns = return_series - rf_series
    mean_excess_return = excess_returns.mean() * N
    volatility = excess_returns.std() * np.sqrt(N)
    return mean_excess_return / volatility


# Sharpe ratio by year
sharpes_by_year = returns.groupby(returns.index.year).apply(
    lambda x: pd.Series({
        'Gold': sharpe_ratio(x['Gold'], x['USD 10y yield'], len(x)),
        'Silver': sharpe_ratio(x['Silver'], x['USD 10y yield'], len(x)) #dynamic N to avoid problem with the last year
    })
)

# Total Sharpe ratio
sharpes_global = pd.Series({
    'Gold': sharpe_ratio(returns['Gold'], returns['USD 10y yield'], len(returns)),
    'Silver': sharpe_ratio(returns['Silver'], returns['USD 10y yield'], len(returns))
})
```

```python
# Plot Sharpe Ratios by year
sharpes_by_year.plot.bar(figsize=(10, 6))
plt.title("Sharpe Ratios by Year for Gold and Silver")
plt.ylabel("Sharpe Ratio")
plt.xlabel("Year")
plt.show()
```

```python
import seaborn as sns
from statsmodels.tsa.stattools import adfuller

def adf_test(series):
    series = series.dropna()  # Without NaN value
    if len(series) < 10: 
        return {"Test Statistic": np.nan, "p-value": np.nan, "Stationary": np.nan}
    result = adfuller(series, autolag="AIC")
    return {
        "Test Statistic": result[0],
        "p-value": result[1],
        "Critical Value (1%)": result[4]["1%"],
        "Critical Value (5%)": result[4]["5%"],
        "Critical Value (10%)": result[4]["10%"],
        "Stationary": result[1] < 0.05,
    }

# Apply the test for each commodity
adf_results_returns = {col: adf_test(commodity[col]) for col in commodity.columns}

adf_df_returns = pd.DataFrame(adf_results_returns).T

plt.figure(figsize=(10, 6))
sns.barplot(x=adf_df_returns.index, y=adf_df_returns["Test Statistic"], palette="coolwarm")
plt.axhline(y=adf_df_returns["Critical Value (5%)"].mean(), color="red", linestyle="dashed", label="Critical Value (5%)")
plt.ylabel("ADF Test Statistic")
plt.title("ADF Test Results for Price")
plt.xticks(rotation=90)
plt.legend()
plt.show()

adf_df_returns
```

For each commodity, compute its 21 day - rolling realized volatility. Display in scatterplots realized vols vs prices. Comment.  



```python
# Calculate 21-day rolling realized volatility for each commodity
rolling_vol = returns_commodity.rolling(window=21).std() * np.sqrt(252)  # Annualized

# Create a function to plot volatility vs price for each commodity
def plot_vol_vs_price(commodity_name):
    plt.figure(figsize=(10, 6))
    plt.scatter(commodity[commodity_name], rolling_vol[commodity_name], alpha=0.5)
    plt.title(f'Price vs Realized Volatility: {commodity_name}')
    plt.xlabel('Price')
    plt.ylabel('21-day Rolling Realized Volatility (Annualized)')
    plt.grid(True, alpha=0.3)
    # Add regression line
    valid_data = ~(np.isnan(commodity[commodity_name]) | np.isnan(rolling_vol[commodity_name]))
    if sum(valid_data) > 2:  # Need at least two points for regression
        x = commodity[commodity_name][valid_data]
        y = rolling_vol[commodity_name][valid_data]
        z = np.polyfit(x, y, 1)
        p = np.poly1d(z)
        plt.plot(x, p(x), 'r--', alpha=0.8)
        # Calculate correlation
        corr = np.corrcoef(x, y)[0, 1]
        plt.annotate(f'Correlation: {corr:.2f}', xy=(0.05, 0.95), xycoords='axes fraction')
    plt.tight_layout()
    plt.show()

# Plot for each commodity (first 4 for brevity)
for commodity_name in list(commodity.columns)[:4]:
    plot_vol_vs_price(commodity_name)
```

```python
# Create a heatmap to visualize the relationship between volatility and prices for all commodities
correlations = []
commodities = []

for commodity_name in commodity.columns:
    valid_data = ~(np.isnan(commodity[commodity_name]) | np.isnan(rolling_vol[commodity_name]))
    if sum(valid_data) > 2:  # Need at least two points for correlation
        x = commodity[commodity_name][valid_data]
        y = rolling_vol[commodity_name][valid_data]
        corr = np.corrcoef(x, y)[0, 1]
        correlations.append(corr)
        commodities.append(commodity_name)

# Create a DataFrame for the correlations
corr_df = pd.DataFrame({'Commodity': commodities, 'Correlation': correlations})
corr_df = corr_df.sort_values('Correlation')

# Plot the correlations
plt.figure(figsize=(10, 6))
sns.barplot(x='Correlation', y='Commodity', data=corr_df, palette='coolwarm')
plt.title('Correlation between Price and Volatility for Each Commodity')
plt.axvline(x=0, color='black', linestyle='--')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

# Comment on the findings
print("Analysis of Price vs. Volatility Relationships:")
print("1. Most commodities show some level of correlation between price and volatility")
print("2. Positive correlation suggests that higher prices are associated with higher volatility")
print("3. Negative correlation suggests that lower prices are associated with higher volatility (volatility clustering during bearish periods)")
print("4. This relationship is commonly observed in commodities and is often referred to as the 'leverage effect'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Load demand data
demand_data = pd.read_csv("demand.csv")

# Convert the year column to datetime for easier manipulation
demand_data['Year'] = pd.to_datetime(demand_data['Dates'].astype(str))

# Resample prices to yearly averages for Chicago wheat and soybeans
yearly_prices = commodity[['Wheat Chicago', 'Soybean']].resample('Y').mean()
yearly_prices.index = yearly_prices.index.year
yearly_prices = yearly_prices.reset_index()
yearly_prices.rename(columns={'index': 'Year'}, inplace=True)
yearly_prices['Date'] = pd.to_datetime(yearly_prices['Date'].astype(str))

# Rename the column 'Dates' to 'Date' in demand_data and ensure 'Date' is datetime
demand_data.rename(columns={'Dates': 'Date'}, inplace=True)
demand_data['Date'] = pd.to_datetime(demand_data['Date'])

# Adjust the dates in demand_data['Date'] to be January 1st of the following year
demand_data['Date'] = demand_data['Date'] + pd.offsets.YearBegin(1)

# Merge dataframes
merged_data = pd.merge(demand_data, yearly_prices, on='Date', how='inner')

# Display the first few rows of the merged data
merged_data[['Year', 'Wheat_United States', 'Wheat Chicago', 'Soybean', 'US_population', 'US_GDP']].head()
```

```python
# Log transform the variables for elasticity analysis
merged_data['log_wheat_demand_us'] = np.log(merged_data['Wheat_United States'])
merged_data['log_wheat_price'] = np.log(merged_data['CBOT Wheat'])

# Create a scatter plot to visualize relationship
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['log_wheat_price'], merged_data['log_wheat_demand_us'], alpha=0.7)
plt.title('Log Wheat Price vs Log Wheat Demand (US)')
plt.xlabel('Log Price')
plt.ylabel('Log Demand')
plt.grid(True, alpha=0.3)

# Add regression line
z = np.polyfit(merged_data['log_wheat_price'], merged_data['log_wheat_demand_us'], 1)
p = np.poly1d(z)
plt.plot(merged_data['log_wheat_price'], p(merged_data['log_wheat_price']), "r--", alpha=0.8)
plt.annotate(f'Slope: {z[0]:.3f}', xy=(0.05, 0.95), xycoords='axes fraction')
plt.show()

# Run OLS regression to estimate price elasticity of wheat demand
X = merged_data['log_wheat_price']
X = sm.add_constant(X)
Y = merged_data['log_wheat_demand_us']

model = sm.OLS(Y, X).fit()
print(model.summary())

print("\nInterpretation of Wheat Price Elasticity:")
elasticity = model.params['log_wheat_price']
print(f"The estimated price elasticity of wheat demand is {elasticity:.4f}")

if abs(elasticity) < 1:
    print("This indicates inelastic demand: changes in price have a relatively small effect on quantity demanded.")
    print("This is typical for staple food products like wheat, which are considered necessities.")
else:
    print("This indicates elastic demand: changes in price have a relatively large effect on quantity demanded.")

# Calculate R-squared to assess model fit
print(f"R-squared: {model.rsquared:.4f}")
print(f"The model explains {model.rsquared*100:.2f}% of the variation in wheat demand")
```

### Question 8
For each of the selected countries, compute (using bivariate regressions):

```python
# Select countries for analysis 
countries = ['United States', 'China', 'Brazil', 'Russia', 'India', 'EU-25', 'Argentina', 'Canada']

# Function to run bivariate analysis for a country
def analyze_elasticities(country):
    # Prepare data variables
    log_wheat_demand = np.log(merged_data[f'Wheat_{country}'])
    log_soy_demand = np.log(merged_data[f'Soybean_{country}'])
    log_wheat_price = merged_data['log_wheat_price']
    log_soy_price = merged_data['log_soybean_price']
    
    # Create matrix of explanatory variables
    X = sm.add_constant(pd.DataFrame({
        'log_wheat_price': log_wheat_price,
        'log_soybean_price': log_soy_price
    }))
    
    # Run wheat demand model
    try:
        wheat_model = sm.OLS(log_wheat_demand, X).fit()
        wheat_own_elasticity = wheat_model.params['log_wheat_price']
        wheat_cross_elasticity = wheat_model.params['log_soybean_price']
        wheat_r2 = wheat_model.rsquared
    except:
        wheat_own_elasticity = wheat_cross_elasticity = wheat_r2 = np.nan
    
    # Run soybean demand model
    try:
        soy_model = sm.OLS(log_soy_demand, X).fit()
        soy_own_elasticity = soy_model.params['log_soybean_price']
        soy_cross_elasticity = soy_model.params['log_wheat_price']
        soy_r2 = soy_model.rsquared
    except:
        soy_own_elasticity = soy_cross_elasticity = soy_r2 = np.nan
    
    return {
        'Country': country,
        'Wheat Own-Price Elasticity': wheat_own_elasticity,
        'Wheat Cross-Price Elasticity': wheat_cross_elasticity,
        'Wheat RÂ²': wheat_r2,
        'Soybean Own-Price Elasticity': soy_own_elasticity,
        'Soybean Cross-Price Elasticity': soy_cross_elasticity,
        'Soybean RÂ²': soy_r2,
        'Wheat-Soy Relationship': 'Substitutes' if wheat_cross_elasticity > 0 else 'Complements' if wheat_cross_elasticity < 0 else 'Independent'
    }

# Apply the function to all selected countries
results = []
for country in countries:
    try:
        result = analyze_elasticities(country)
        results.append(result)
        print(f"Completed analysis for {country}")
    except Exception as e:
        print(f"Error analyzing {country}: {e}")

# Create results dataframe
elasticity_results = pd.DataFrame(results)
elasticity_results
```

### Question 10

Make sure that the prices for Wheat in Chicago, Minneapolis and MATIF are expressed in the same currency and unit. 
1. Plot the time series of the spreads Chicago/Matif and Chicago/Minneapolis
2. Are those mean-reverting? (show evidence!)

```python
# First, ensure all wheat prices are in the same currency and unit
# Identify the wheat price columns
wheat_cols = ['CBOT Wheat', 'MGEX Wheat', 'MATIF Wheat']

# Check if all wheat prices are in the data
print("Wheat price columns available in our dataset:")
for col in wheat_cols:
    print(f"- {col}: {'Present' if col in data.columns else 'Not found'}")

# Make a copy of the data for analysis
wheat_data = data[wheat_cols].copy()

# Convert MATIF wheat price to USD if it's in EUR
if 'MATIF Wheat' in wheat_data.columns and 'EURUSD' in data.columns:
    # Check the currency from the Currency dataframe
    is_eur = False
    for col in range(Currency.shape[1]):
        if 'MATIF Wheat' in Currency.columns[col] and 'EUR' in Currency.iloc[0, col]:
            is_eur = True
            break
    
    if is_eur:
        print("\nConverting MATIF Wheat from EUR to USD...")
        wheat_data['MATIF Wheat'] = wheat_data['MATIF Wheat'] * data['EURUSD']
        print("Conversion completed.")
    else:
        print("\nMATIF Wheat is already in USD, no conversion needed.")

# 1. Plot the time series of the spreads
plt.figure(figsize=(14, 8))

# Chicago/MATIF spread
plt.subplot(2, 1, 1)
spread_chicago_matif = wheat_data['CBOT Wheat'] - wheat_data['MATIF Wheat']
plt.plot(spread_chicago_matif.index, spread_chicago_matif, 'b-')
plt.title('CBOT/MATIF Wheat Price Spread')
plt.ylabel('Price Spread (USD)')
plt.grid(True, alpha=0.3)
plt.axhline(y=spread_chicago_matif.mean(), color='r', linestyle='--', 
           label=f'Mean: {spread_chicago_matif.mean():.2f}')
plt.legend()

# Chicago/Minneapolis spread
plt.subplot(2, 1, 2)
spread_chicago_mpls = wheat_data['CBOT Wheat'] - wheat_data['MGEX Wheat']
plt.plot(spread_chicago_mpls.index, spread_chicago_mpls, 'g-')
plt.title('CBOT/MGEX Wheat Price Spread')
plt.xlabel('Date')
plt.ylabel('Price Spread (USD)')
plt.grid(True, alpha=0.3)
plt.axhline(y=spread_chicago_mpls.mean(), color='r', linestyle='--', 
           label=f'Mean: {spread_chicago_mpls.mean():.2f}')
plt.legend()

plt.tight_layout()
plt.show()

# 2. Test for mean reversion using Augmented Dickey-Fuller test
from statsmodels.tsa.stattools import adfuller

def test_mean_reversion(series, name):
    # Drop NaNs
    series = series.dropna()
    
    # Run ADF test
    result = adfuller(series, autolag='AIC')
    
    print(f"Mean Reversion Test for {name}:")
    print(f"ADF Statistic: {result[0]:.4f}")
    print(f"p-value: {result[1]:.4f}")
    
    # Print critical values
    print("Critical Values:")
    for key, value in result[4].items():
        print(f"   {key}: {value:.4f}")
    
    # Interpret result
    if result[1] < 0.05:
        print(f"Result: The {name} spread is stationary/mean-reverting (reject unit root)")
    else:
        print(f"Result: The {name} spread is not stationary/mean-reverting (cannot reject unit root)")
    print("\n")
    
    return result

# Test both spreads
test_chicago_matif = test_mean_reversion(spread_chicago_matif, "CBOT/MATIF")
test_chicago_mpls = test_mean_reversion(spread_chicago_mpls, "CBOT/MGEX")

# 3. Further analysis: Half-life of mean reversion
def estimate_half_life(spread):
    # Calculate lag-1 spread
    spread_lag = spread.shift(1)
    
    # Remove NAs
    spread = spread[1:]
    spread_lag = spread_lag[1:]
    
    # Run regression for AR(1) coefficient
    reg = sm.OLS(spread - spread_lag, spread_lag).fit()
    ar1_coef = -reg.params[0]
    
    # Calculate half-life
    half_life = np.log(2) / ar1_coef if ar1_coef > 0 else np.nan
    
    return half_life

# Calculate half-lives
half_life_chicago_matif = estimate_half_life(spread_chicago_matif)
half_life_chicago_mpls = estimate_half_life(spread_chicago_mpls)

print("\nMean Reversion Half-Life Analysis:")
print(f"CBOT/MATIF Half-Life: {half_life_chicago_matif:.2f} days")
print(f"CBOT/MGEX Half-Life: {half_life_chicago_mpls:.2f} days")

# Visual analysis of mean reversion: Histogram of spreads
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(spread_chicago_matif.dropna(), kde=True, bins=30)
plt.axvline(spread_chicago_matif.mean(), color='r', linestyle='--', label=f'Mean: {spread_chicago_matif.mean():.2f}')
plt.title('Distribution of CBOT/MATIF Spread')
plt.xlabel('Spread Value')
plt.legend()

plt.subplot(1, 2, 2)
sns.histplot(spread_chicago_mpls.dropna(), kde=True, bins=30)
plt.axvline(spread_chicago_mpls.mean(), color='r', linestyle='--', label=f'Mean: {spread_chicago_mpls.mean():.2f}')
plt.title('Distribution of CBOT/MGEX Spread')
plt.xlabel('Spread Value')
plt.legend()

plt.tight_layout()
plt.show()

# Interpretation summary
print("\nInterpretation of Wheat Spread Analysis:")
print("1. Geographic Arbitrage: The spreads between wheat markets reflect transportation costs,")
print("   quality differences, and local supply-demand dynamics.")
print("2. Mean reversion indicates market efficiency - deviations from the 'normal' spread")
print("   create arbitrage opportunities which traders exploit, pushing prices back toward equilibrium.")
print("3. The half-life metric provides a practical estimate of how quickly these")
print("   arbitrage opportunities disappear, which is critical for trading strategy development.")
print("4. Faster mean reversion (shorter half-life) indicates more efficient markets and")
print("   stronger economic links between the trading locations.")
```

### Question 14

Theoretically, if the mine is not profitable enough, it should shut down. How would you amend the above model to take this into account? Run the same analysis and estimate the parameters. Is this shut-down option materially improving the model?

```python
# Enhanced model with shutdown option
print("Enhancing the gold mine model to account for shutdown option...")

# Theoretical framework:
# - A mine will shut down if the operating margin becomes negative
# - This creates an option-like payoff where the mine's value has a floor
# - We can model this with a max(0, operating_margin) function
# - This nonlinear relationship can be approximated using a piecewise linear model

def model_gold_mine_with_shutdown(miner):
    """
    Build a gold mine model with shutdown option
    Model: Share Price = Option Value / Risk-Free Rate
    where Option Value = max(0, beta_gold * Gold - beta_diesel * Diesel - Fixed Costs)
    """
    # Prepare the data
    df = pd.DataFrame({
        'SharePrice': data[miner],
        'Gold': data['Gold'],
        'Diesel': data['IPE Gasoil'] if 'IPE Gasoil' in data.columns else 0,
        'RiskFreeRate': data['USD 10y yield'] / 100  # Convert from percentage to decimal
    }).dropna()
    
    # Create the basic model variables
    df['GoldOverRate'] = df['Gold'] / df['RiskFreeRate']
    df['DieselOverRate'] = df['Diesel'] / df['RiskFreeRate'] if 'IPE Gasoil' in data.columns else 0
    df['InverseRate'] = 1 / df['RiskFreeRate']
    
    # To approximate the option-like behavior, we'll use a threshold approach
    # Step 1: Fit the basic model first to get initial estimates
    X_basic = sm.add_constant(pd.DataFrame({
        'GoldOverRate': df['GoldOverRate'], 
        'DieselOverRate': df['DieselOverRate'] if 'IPE Gasoil' in data.columns else np.zeros(len(df)),
        'InverseRate': df['InverseRate']
    }))
    Y = df['SharePrice']
    basic_model = sm.OLS(Y, X_basic).fit()
    
    # Extract initial parameters
    beta_gold_init = basic_model.params['GoldOverRate']
    beta_diesel_init = basic_model.params['DieselOverRate'] if 'IPE Gasoil' in data.columns else 0
    fixed_costs_init = -basic_model.params['InverseRate']
    
    # Step 2: Calculate the operating margin based on initial parameters
    df['OperatingMargin'] = beta_gold_init * df['Gold'] - beta_diesel_init * df['Diesel'] - fixed_costs_init
    
    # Step 3: Create a dummy variable for "operating" vs "shutdown" states
    df['IsOperating'] = (df['OperatingMargin'] > 0).astype(int)
    
    # Step 4: Create interaction terms to model different behavior in operating vs shutdown state
    df['GoldOverRate_Op'] = df['GoldOverRate'] * df['IsOperating']
    df['DieselOverRate_Op'] = df['DieselOverRate'] * df['IsOperating'] if 'IPE Gasoil' in data.columns else 0
    df['InverseRate_Op'] = df['InverseRate'] * df['IsOperating']
    
    # Step 5: Fit the enhanced model with shutdown option
    X_shutdown = sm.add_constant(pd.DataFrame({
        'GoldOverRate_Op': df['GoldOverRate_Op'],
        'DieselOverRate_Op': df['DieselOverRate_Op'] if 'IPE Gasoil' in data.columns else np.zeros(len(df)),
        'InverseRate_Op': df['InverseRate_Op'],
        'IsOperating': df['IsOperating']  # Intercept shift for shutdown state
    }))
    
    shutdown_model = sm.OLS(Y, X_shutdown).fit()
    
    # Extract parameters
    beta_gold = shutdown_model.params['GoldOverRate_Op']
    beta_diesel = -shutdown_model.params['DieselOverRate_Op'] if 'IPE Gasoil' in data.columns else 0
    fixed_costs = -shutdown_model.params['InverseRate_Op']
    option_value = shutdown_model.params['IsOperating']  # The value of flexibility
    
    # Calculate predictions
    df['BasicPrediction'] = basic_model.predict(X_basic)
    df['ShutdownPrediction'] = shutdown_model.predict(X_shutdown)
    
    return {
        'Miner': miner,
        'Beta Gold': beta_gold,
        'Beta Diesel': beta_diesel,
        'Fixed Costs': fixed_costs,
        'Option Value': option_value,
        'Basic R-squared': basic_model.rsquared,
        'Shutdown R-squared': shutdown_model.rsquared,
        'Basic Model': basic_model,
        'Shutdown Model': shutdown_model,
        'Data': df,
        'Operating Days %': df['IsOperating'].mean() * 100
    }

# Apply the shutdown model to each gold miner
shutdown_results = []
for miner in gold_miners:
    try:
        result = model_gold_mine_with_shutdown(miner)
        shutdown_results.append(result)
        print(f"\nShutdown Model Results for {miner}:")
        print(f"Beta Gold: {result['Beta Gold']:.4f}")
        if 'IPE Gasoil' in data.columns:
            print(f"Beta Diesel: {result['Beta Diesel']:.4f}")
        print(f"Fixed Costs: {result['Fixed Costs']:.2f}")
        print(f"Option Value: {result['Option Value']:.2f}")
        print(f"Basic Model R-squared: {result['Basic R-squared']:.4f}")
        print(f"Shutdown Model R-squared: {result['Shutdown R-squared']:.4f}")
        print(f"Operating Days: {result['Operating Days %']:.1f}%")
        
        # If there's substantial improvement, print the detailed summary
        if result['Shutdown R-squared'] - result['Basic R-squared'] > 0.05:
            print("\nDetailed Shutdown Model Summary (significant improvement):")
            print(result['Shutdown Model'].summary())
    except Exception as e:
        print(f"Error modeling {miner} with shutdown option: {e}")

# Visualize the results
if shutdown_results:
    plt.figure(figsize=(15, 10))
    
    # Plot comparison of R-squared values
    plt.subplot(2, 2, 1)
    basic_r2 = [r['Basic R-squared'] for r in shutdown_results]
    shutdown_r2 = [r['Shutdown R-squared'] for r in shutdown_results]
    
    x = np.arange(len(gold_miners))
    width = 0.35
    
    plt.bar(x - width/2, basic_r2, width, label='Basic Model', color='lightblue')
    plt.bar(x + width/2, shutdown_r2, width, label='Shutdown Model', color='tomato')
    plt.xticks(x, gold_miners, rotation=90)
    plt.title('Model Performance Comparison')
    plt.ylabel('R-squared')
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    
    # Plot operating days percentage
    plt.subplot(2, 2, 2)
    op_days = [r['Operating Days %'] for r in shutdown_results]
    plt.bar(range(len(gold_miners)), op_days, color='green')
    plt.axhline(y=100, color='red', linestyle='--', label='Full Operation')
    plt.xticks(range(len(gold_miners)), gold_miners, rotation=90)
    plt.title('Percentage of Operating Days by Miner')
    plt.ylabel('Percentage (%)')
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    
    # Plot option value
    plt.subplot(2, 2, 3)
    option_values = [r['Option Value'] for r in shutdown_results]
    plt.bar(range(len(gold_miners)), option_values, color='purple')
    plt.xticks(range(len(gold_miners)), gold_miners, rotation=90)
    plt.title('Shutdown Option Value by Miner')
    plt.ylabel('Option Value')
    plt.grid(axis='y', alpha=0.3)
    
    # Plot fixed costs
    plt.subplot(2, 2, 4)
    fixed_costs = [r['Fixed Costs'] for r in shutdown_results]
    plt.bar(range(len(gold_miners)), fixed_costs, color='gray')
    plt.xticks(range(len(gold_miners)), gold_miners, rotation=90)
    plt.title('Fixed Costs by Miner')
    plt.ylabel('Fixed Costs')
    plt.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Compare the predictions for the first miner
    first_miner = shutdown_results[0]['Miner']
    df = shutdown_results[0]['Data']
    
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['SharePrice'], label='Actual Share Price', color='blue')
    plt.plot(df.index, df['BasicPrediction'], label='Basic Model Prediction', color='red', linestyle='--')
    plt.plot(df.index, df['ShutdownPrediction'], label='Shutdown Model Prediction', color='green', linestyle='-.')
    plt.title(f'Model Comparison with Shutdown Option: {first_miner}')
    plt.xlabel('Date')
    plt.ylabel('Share Price')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Calculate improvement statistics
    improvements = [r['Shutdown R-squared'] - r['Basic R-squared'] for r in shutdown_results]
    avg_improvement = np.mean(improvements)
    max_improvement = np.max(improvements)
    
    print("\nShutdown Option Model Analysis:")
    print(f"Average R-squared improvement: {avg_improvement:.4f} ({avg_improvement*100:.2f}%)")
    print(f"Maximum R-squared improvement: {max_improvement:.4f} ({max_improvement*100:.2f}%)")
    
    print("\nInterpretation of Gold Mine Shutdown Option Model:")
    print("1. The shutdown option represents the flexibility of a mine to temporarily halt")
    print("   operations when the gold price is too low relative to costs.")
    print("2. This creates a floor on losses and changes the relationship between gold price and share price.")
    print("3. Mines with higher fixed costs would benefit more from modeling the shutdown option,")
    print("   as they would be more likely to shut down during adverse price conditions.")
    print("4. The model improvement is more significant for high-cost miners or during periods")
    print("   when gold prices fall below the break-even point for extended periods.")
    print("5. The option value represents the additional value from the flexibility to shut down.")
    
    # Assessing whether the shutdown option materially improves the model
    if avg_improvement > 0.05:
        print("\nConclusion: The shutdown option MATERIALLY improves the model fit.")
        print("This suggests that investors do value the flexibility of mines to shut down in adverse conditions.")
    else:
        print("\nConclusion: The shutdown option does not materially improve the model fit.")
        print("This could be because:")
        print("a) Gold prices mostly stayed above break-even points during the sample period")
        print("b) Investors may not fully price in the shutdown option")
        print("c) Our piecewise linear approximation may not fully capture the option characteristics")
```
